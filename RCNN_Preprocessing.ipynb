```python
import os,cv2,keras
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import tensorflow as tf
import random
from PIL import Image
```

```python
df = pd.read_csv("train.csv")
```

```python
def bb_intersection_over_union(boxA, boxB):
    # determine the (x, y)-coordinates of the intersection rectangle
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    # compute the area of intersection rectangle
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)
    # compute the area of both the prediction and ground-truth
    # rectangles
    boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)
        # compute the intersection over union by taking the intersection
        # area and dividing it by the sum of prediction + ground-truth
        # areas - the interesection area
    iou = interArea / float(boxAArea + boxBArea - interArea)
    # return the intersection over union value
#     print("value : "+str(iou))
    return iou
```

```python
train_images=[]
train_labels=[]
```

```python
#preprocessing
for index, row in df.iterrows():
    try:
        image_name = "images/"+row[0]
#         print(image_name)
        x = row[1]
        y = row[2]
        w = row[3]
        h = row[4]
        # print(row)
        image = cv2.imread(image_name)
        image_name = row[0]
        image_name = image_name[:-4]
        # plt.imshow(image)
        gtvalues = []
        gtvalues.append([x, y, w, h])
        i=0
        results = []
        while i<2000:
            i+=1
            x1 = random.randint(200,500)
            y1 = random.randint(200,500)
            w1 = random.randint(1500,2500)
            h1 = random.randint(1000,3000)
            results.append([x1,y1,w1,h1])
        results.append([x,y,w,h])  
#         print(results)
        imout = image.copy()
        counter = 0
        falsecounter = 0
        flag = 0
        fflag = 0
        bflag = 0
        for e, result in enumerate(results):
#             print(e)
            if e < 2000 and flag == 0:
                for gtval in gtvalues:
                    x,y,w,h = result
                    iou = bb_intersection_over_union(gtval,[x,y,w,h])
                    if counter < 30:
                        if iou > 0.60:
                            timage = imout[y:y+h,x:x+w]
                            resized = cv2.resize(timage, (224,224), interpolation = cv2.INTER_AREA)
                            train_images.append(resized)
                            train_labels.append(1)
                            counter += 1
                            np.save("processed/"+image_name+str(e), resized)
                            print(str(e)+" : 1 ; iou = "+ str(iou))
                        else :
                            fflag =1
                        if falsecounter <30:
                            if iou < 0.3:
                                timage = imout[y:y+h,x:x+w]
                                resized = cv2.resize(timage, (224,224), interpolation = cv2.INTER_AREA)
                                train_images.append(resized)
                                train_labels.append(0)
                                falsecounter += 1
                                np.save("processed/"+image_name+str(e), resized)
                                print(str(e)+" : 0 ; iou = "+ str(iou))
                        else :
                            bflag = 1
                if fflag == 1 and bflag == 1:
                    print("inside")
                    flag = 1
    except Exception as e:
        print(e)
        print("error in image/"+row[0])
        continue
```
